```python
#Using all features to build a model
X = df.drop(['id','stroke','gender','ever_married','work_type','Residence_type','smoking_status'],axis=1)
Y = df['stroke']
```

```python
#Normalize data to common scale
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(X.values)
scaled_values = scale.transform(X.values)
```

```python
scaled_values
```

```python
#New dataframe with scaled values
df1 = pd.DataFrame(data=scaled_values,columns=x.columns)
df1.head()
```

# 1. K-Nearest Neighbor

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1, Y, test_size=0.3,random_state=106)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
prediction = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
print(confusion_matrix(y_test,prediction))
```

```python
print(classification_report(y_test,prediction))
```

```python
print("Accuracy from KNN: ",accuracy_score(y_test,prediction))
```

# 2. Naive Bais

```python
feature_drop = []
```

```python
# Using important features only (manual selection)
X1 = df1.drop(feature_drop,axis=1)
Y1 = df['stroke']
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, Y1, test_size = 0.3, random_state=106)
```

```python
X1
```

```python
from sklearn.naive_bayes import GaussianNB
gauss_cl = GaussianNB()
gauss_cl.fit(X1_train,y1_train)
```

```python
gauss_pred = gauss_cl.predict(X1_test)
```

```python
print(confusion_matrix(y1_test,gauss_pred))
print('\n')
print(classification_report(y1_test,gauss_pred))
print('\n')
print("Accuracy from Naive Bayes: ",accuracy_score(y1_test,gauss_pred))
```

# 3. Random Forest

```python
#Before feature selection
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50)
rfc.fit(X_train,y_train)
```

```python
prediction1 = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,prediction1))
print('\n')
print(classification_report(y_test,prediction1))
print('\n')
print("Accuracy from Random Forest before using feature selection: ",accuracy_score(y1_test,prediction1))
```

```python
#Using Recursive feature elimination with cross validation
from sklearn.feature_selection import RFECV
rfecv_cl = RFECV(estimator=rfc, step=1, cv=5,scoring='accuracy')   
rfecv_cl = rfecv_cl.fit(X_train, y_train)
```

```python
print('Optimal number of features :', rfecv_cl.n_features_)
print('Best features :', X_train.columns[rfecv_cl.support_])
```

```python
# plt.figure(figsize=(12, 6))
# plt.title('Recursive Feature Elimination with Cross-Validation', fontsize=15, fontweight='bold')
# plt.xlabel('Number of features selected', fontsize=12)
# plt.ylabel('% Accuracy', fontsize=12)
# plt.plot(range(1, len(rfecv_cl.grid_scores_) + 1), rfecv_cl.grid_scores_, lw=2)
# plt.show()

plt.figure(figsize=(12, 6))
plt.title('Recursive Feature Elimination with Cross-Validation', fontsize=15, fontweight='bold')
plt.xlabel('Number of features selected', fontsize=12)
plt.ylabel('% Accuracy', fontsize=12)
plt.plot(range(1, len(rfecv_cl.grid_scores_) + 1), rfecv_cl.cv_scores_, lw=2)
plt.show()
```

```python
rfecv_pred = rfecv_cl.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfecv_pred))
print('\n')
print(classification_report(y_test,rfecv_pred))
print('\n')
print("Accuracy from Random Forest after using feature selection: ",accuracy_score(y_test,rfecv_pred))
```
